# 03 Opening a csv file from an URL

```python
# Importing the 'requests' module to send HTTP requests
# HTTP request bhejne ke liye 'requests' module import kar rahe hain
import requests

# Importing StringIO to convert text data into a file-like object
# StringIO ka use text ko file jaisa object banane ke liye hota hai
from io import StringIO

# URL of the CSV file (this is a publicly available CSV file from GitHub)
# CSV file ka URL (ye ek public GitHub par available CSV file hai)
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"

# Some websites block bots, so we add a User-Agent header to mimic a browser request
# Kai websites bot traffic block karti hain, isliye User-Agent header bhejna padta hai taki request browser jaisi lage
header = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"
}

# Sending a GET request to the URL and storing the response in 'req'
# GET request bhej rahe hain aur response ko 'req' variable me store kar rahe hain
req = requests.get(url, headers=header)

# Converting the text content of the response into a file-like object using StringIO
# Response ka text (CSV content) ko StringIO ke through file-like object me convert kar rahe hain
data = StringIO(req.text)

# Reading the CSV data using pandas and converting it into a DataFrame
# Pandas ka read_csv() function use karke text-based CSV ko DataFrame me badal rahe hain
import pandas as pd
pd.read_csv(data)
```

```python
# Reading only the 'gender' column from the CSV file and converting it into a Series using squeeze()
# CSV file me se sirf 'gender' column read kar rahe hain, aur squeeze() function se usse Series ya scalar me convert kar rahe hain

pd.read_csv(aug_train, usecols=['gender']).squeeze()

# usecols=['gender']: Iska matlab sirf 'gender' column ko read karo
# squeeze(): agar DataFrame me sirf ek column ya row ho to usse squeeze karke Series (ya agar ek hi value ho to scalar) me badal deta hai
# Agar column me ek se zyada rows hain to result hamesha Series hoga, scalar tabhi hota hai jab sirf ek value ho
```

```python
# Reading only the first 100 rows from the CSV file using nrows parameter
# nrows parameter ka use karke CSV file me se sirf pehli 100 rows ko read kar rahe hain

pd.read_csv(aug_train, nrows=100)

# nrows=100: iska matlab hai ki file me se sirf top (shuru ki) 100 rows read ki jayengi
# Yeh kaafi useful hota hai jab aap large dataset ka preview lena chahte hain bina pura load kiye
```

```python
# Reading the Zomato dataset with proper encoding to avoid UnicodeDecodeError
# UnicodeDecodeError ko avoid karne ke liye Zomato dataset ko sahi encoding ke saath read kar rahe hain

df = pd.read_csv(zomato, encoding='latin-1')
df

# Jab CSV file 'utf-8' encoding me nahi hoti aur aap use utf-8 me read karte hain, to UnicodeDecodeError aata hai
# encoding='latin-1' ek alternative encoding hai jo aise special characters (€, ñ, é etc.) ko support karti hai
# Ye approach mostly tab use hoti hai jab data India ke bahar ya multi-language source se ho

# df ka matlab hai DataFrame — jisme Zomato ka pura data ab load ho chuka hai
```

```python
# 11 Skip bad lines
pd.read_csv(BX_books,encoding='latin-1')
```

```python
# Reading the BX_books dataset with proper separator, encoding, and handling bad lines
# BX_books dataset ko correct separator (';'), encoding ('latin-1') aur format mismatch rows ko handle karte hue read kar rahe hain

pd.read_csv(BX_books, sep=';', encoding='latin-1', on_bad_lines='skip')

# sep=';' : CSV file me columns semicolon (;) se separate hain, isliye separator define kiya gaya hai
# encoding='latin-1' : Special characters (like ñ, é, ü etc.) handle karne ke liye
# on_bad_lines='skip' : Aisi lines jinka format sahi nahi hai (jaise extra ya missing column values), unhe skip kar diya jata hai

# Agar ye parameter na ho to pandas error throw karega jab koi badly formatted line milegi
```

# Handling Dates

```python
# Reading the IPL dataset and displaying column info using .info()
# IPL dataset ko read kar rahe hain aur .info() se column details jaane rahe hain

pd.read_csv(IPL).info()

# Yahan par date wala column 'object' type me show ho raha hai
# iska matlab hai ki pandas ne usse string ke form me read kiya hai
# jabki ideally wo 'datetime64' hona chahiye taaki date-based operations apply ho saken
```

```python
# Reading the IPL dataset and parsing the 'date' column as datetime format
# IPL dataset ko read karte waqt 'date' column ko datetime format me convert kar rahe hain

pd.read_csv(IPL, parse_dates=['date']).info()

# parse_dates=['date']:
# Pandas ko batata hai ki 'date' column ko string (object) ke bajaye datetime64[ns] format me parse karo
# Iska fayda ye hai ki baad me aap date-based filtering, year/month extraction, sorting etc. asaani se kar sakte ho

# .info():
# Ye function DataFrame ke structure ko show karta hai, jisme column names, non-null values aur data types dikhte hain

# 🔍 Output me ab 'date' column ka dtype hoga: datetime64[ns] (na ki object)
```
